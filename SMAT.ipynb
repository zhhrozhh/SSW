```python
X = []
Y = []
Z = []
for i in range(-5,5):
    for j in range(-5,5):
        for k in range(-5,5):
            if i+j+k == 2:
                X.append(i)
                Y.append(j)
                Z.append(k)
```

```python
w = list(zip(X,Y,Z))
```

```python
W = np.array(
    [
        [1/np.sqrt(2),-1/np.sqrt(2),0],
        [1/np.sqrt(6),1/np.sqrt(6),-2/np.sqrt(6)],
        [1/np.sqrt(3) for i in range(3)]
    ]
)
```

```python
def G(x):
    return np.array(
        [
            [np.cos(x),np.sin(x),0],
            [-np.sin(x),np.cos(x),0],
            [0,0,1]
        ]
    )
```

```python
def R(x):
    return np.matmul(np.matmul(W.T,G(x)),W)
```

```python
XX = [[1,1,-2]]
r = R(0.03)
for i in range(10000):
    XX.append(np.matmul(r,XX[-1]))
xd = []
yd = []
zd = []
for s in XX:
    xd.append(s[0])
    yd.append(s[1])
    zd.append(s[2])
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xd,yd,zd)
```
